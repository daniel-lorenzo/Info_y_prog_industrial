## Ejercicio 2
Un motor eléctrico puede girar en ambos sentidos por medio de dos contactores: “D” para el giro a derecha e “I” para el giro a izquierda. Estos dos contactores son comandados por dos pulsadores de giro “d” (derecha) e “i” (izquierda) y un interruptor de selección “L” de acuerdo con las siguientes condiciones:
- Si sólo se pulsa uno de los dos botones de giro, el motor gira en el sentido correspondiente.
- Si se pulsan los dos botones de giro simultáneamente, el sentido de giro depende del estado del interruptor “L” de forma que,
	* Si “L” está activado, el motor gira a la derecha
	* Si “L” está en reposo, el motor gira a la izquerda

Establecer:
1. La tabla de verdad
2. Las funciones lógicas D e I y simplificarlar
3. Su circuito lógico mediante compuertas

### Solución
**Tabla de verdad**

imagen.png

**Mapa de Karnaugh** $F(I)$

imagen.png

$$\begin{align*}
I &= i.d' + L'.i \\
I &= i.(d' + L') \\
I &= i.(\overline{d.L})\\
\end{align*}$$

**Mapa de Karnaugh** $F(D)$

imagen.png

$$\begin{align*}
D &= i'.d + d.L \\
D &= d.(i' + L)
\end{align*}$$

## Ejercicio 10
Se desea controlar una lámpara empleando tres interruptores, de forma que sólo se encienda cuando esté activado un solo interruptor o los tres simultáneamente. Se pide:

	a. La tabla de verdad
	b. La función lógica
	c. Realizar un circuito con compuertas lógicas que lo ejecute

### Solución
**Tabla de verdad**

imagen.png

**Mapa de Karnaugh**

imagen.png

$L = CA'B' + C'A'B + CAB + C'AB'$

Aplicamos algebra de Boole:

$$\begin{align*}
 L &= CA'B' + C'A'B + CAB + C'AB' \\
 L &= CA'B' + C'A'B + CAB + C'AB' + CA'A + CB'B \\
 L &= C'( A'B + AB' ) + C(A'A + AB ) + C( A'B' + B'B ) \\
 L &= C' (A \oplus B ) + CA(A'+B) + CB' (A'+B') \\
 L &= C'(A \oplus B) + C(A' + B)(A+B') \\
 L &= C'(A \oplus B) + C(\overline{A' B}) (\overline{AB'}) \\
 L &= C'(A \oplus B) + C(\overline{A' B + AB'}) \\
 L &= C'(A \oplus B) + C(\overline{A \oplus B})
 \end{align*}$$

Entonces:
$$ L = C \oplus ( A  \oplus B) $$

## Ejercicio 12
Diseñar un sumador aritmético de un bit, con la capacidad de conectarse en cascada para ampliar la cantidad de bits a sumar.

### Solución
**Tabla de verdad**

image.png

**Mapa de Karnaugh**

image.png

$$\begin{align*}
c_{out} &= x.y + c_{in}.x + c_{in}.y \\
c_{out} &= c_{in}.(x+y) + xy
\end{align*}$$

image.png

$$ s = c_{in}.x'.y' + c_{in}' x'.y + c_{in}' xy' + c_{in}.x.y $$

Agregamos los términos núlos:

$$ s = c_{in}.x'.y' + c_{in}' x'.y + c_{in}' xy' + c_{in}.x.y + c_{in}.x'.x + c_{in}.y'.y $$

Agrupamos términos factorizando:

$$ s = c_{in}'.(x'.y + x.y') + c_{in}.(x.x' + x.y) + c_{in}.(x'.y' + y'.y) $$

Sacando factor común:

$$\begin{align*}
s &= c_{in}'.(x'.y + x.y') + c_{in}.x.(x' + y) + c_{in}.y'(x' + y) \\
s &= c_{in}'.(x \oplus y) + c_{in}.(x' + y)(x + y') \\
s &= c_{in}'.(x \oplus y) + c_{in}.(\overline{x'.y})(\overline{x.y'}) \\
s &= c'_{in}.(x \oplus y) + c_{in}.(\overline{x'.y + x.y'}) \\
s &= c_{in}'.(x \oplus y) + c_{in}.(\overline{x \oplus y}) \\
s &= c_{in} \oplus (x \oplus y)
\end{align*}$$
